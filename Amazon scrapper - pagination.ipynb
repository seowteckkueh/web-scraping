```python
from requests_html import HTMLSession
from bs4 import BeautifulSoup
import time

s = HTMLSession()

url='https://www.amazon.co.uk/s?k=dslr+camera'

def getdata(url):
    r = s.get(url)
    #r.html.render(sleep=1)
    # StackOverflow: jupyternotebook has it's own event timer, hence this can only be run with asyncio
    # or directly from IDE
    soup=BeautifulSoup(r.html.html, 'html.parser')
    return soup

def getnextpage(soup):
    page=soup.find('span',{'data-component-type':'s-search-results'})
    if not page.find('span',{'class':'s-pagination-item s-pagination-next s-pagination-disabled'}):
        next=page.find('a',{'class':'s-pagination-item s-pagination-next s-pagination-button s-pagination-separator'})['href']
        next_url='https://www.amazon.co.uk'+str(next)
        return next_url
    else:
        return
url_list=[]
while True:
    soup=getdata(url)
    new_url=getnextpage(soup)
    url_list.append(new_url)
    if not new_url:
        break
```
